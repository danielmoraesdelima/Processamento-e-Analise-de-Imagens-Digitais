```python
# Funções Originais do Prof. Daniel enviada no notebook original

def translate_image(image, translation):
  new_image = np.zeros(image.shape, dtype='uint8')

  #completar    
  h,w = image.shape[:2]
  x_distance = translation[0]
  y_distance = translation[1]
  ts_mat = np.array([[1,0,x_distance],[0,1,y_distance]])

  #dica
  for i in range(0, image.shape[1]): #W
    for j in range(0, image.shape[0]): #H
      origin_x = j
      origin_y = i
      origin_xy = np.array([origin_x,origin_y,1])
            
      new_xy = np.dot(ts_mat,origin_xy)
      new_x = new_xy[0]
      new_y = new_xy[1]

      if 0<new_x < w and 0<new_y < h:
        new_image[new_y,new_x] = image[i,j]

  return new_image

def rotate_image(image, angle, pivot_point):
  new_image = np.zeros(image.shape, dtype='uint8')

  #completar  
  #1.create rotation matrix with numpy array
  rotation_mat = np.transpose(np.array([[np.cos(angle),-np.sin(angle)],
                                        [np.sin(angle),np.cos(angle)]]))
  h,w = image.shape[:2]
    
  pivot_point_x =  pivot_point[0]
  pivot_point_y = pivot_point[1]
    
  for height in range(h): #h = number of row
    for width in range(w): #w = number of col
      xy_mat = np.array([[width-pivot_point_x],[height-pivot_point_y]])
            
      rotate_mat = np.dot(rotation_mat,xy_mat)

      new_x = pivot_point_x + int(rotate_mat[0])
      new_y = pivot_point_y + int(rotate_mat[1])

      if (0<=new_x<=w-1) and (0<=new_y<=h-1): 
        new_image[new_y,new_x] = image[height,width]        

  return new_image

def scale_image(image, scale):
  new_image = np.zeros(image.shape, dtype='uint8')

  #completar 
  W, H = image.shape[:2]
  new_W, new_H = scale[0],scale[1]
  xrange = lambda x: np.linspace(0, 1, x)
  f = interpolate.interp2d(xrange(W), xrange(H), image, kind="linear")
  new_image = f(xrange(new_W), xrange(new_H))

  return new_image

def rotate_image_from_center(image, angle, pivot_point):
  new_image = np.zeros(image.shape, dtype='uint8')

  #completar  
  #1.create rotation matrix with numpy array
  rotation_mat = np.transpose(np.array([[np.cos(angle),-np.sin(angle)],
                                        [np.sin(angle),np.cos(angle)]]))
  h,w = image.shape[:2]
    
  pivot_point_x =  pivot_point[0]
  pivot_point_y = pivot_point[1]
    
  for height in range(h): #h = number of row
    for width in range(w): #w = number of col
      xy_mat = np.array([[width-pivot_point_x],[height-pivot_point_y]])
            
      rotate_mat = np.dot(rotation_mat,xy_mat)

      new_x = pivot_point_x + int(rotate_mat[0])
      new_y = pivot_point_y + int(rotate_mat[1])

      if (0<=new_x<=w-1) and (0<=new_y<=h-1): 
        new_image[new_y,new_x] = image[height,width]        

  return new_image

def affine_transform_image(image, matrix):
  new_image = np.zeros(image.shape, dtype='uint8')

  #completar         

  return new_image
```

```python
#Funcao para Exibir Imagens com matplotlib e cv2
def exibir_imgplt(tmpimg):
  #Importar a Imagem
  image = cv2.imread(tmpimg)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  image = cv2.resize(image,None,fx=0.25,fy=0.25,interpolation=cv2.INTER_AREA)
  #Mostra a imagem com matplotlib
  plt.imshow(image)
  plt.show()
  return
```
