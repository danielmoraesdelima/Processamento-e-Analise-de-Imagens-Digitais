#### Funções

```python
#Funcao para Exibir Imagens com matplotlib e cv2
def exibir_imgplt(tmpimg):
  #Importar a Imagem
  image = cv2.imread(tmpimg)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  image = cv2.resize(image,None,fx=0.25,fy=0.25,interpolation=cv2.INTER_AREA)
  #Mostra a imagem com matplotlib
  plt.imshow(image)
  plt.show()
  return


#Função para Exibir a Imagem utiliznado OpenCV com a função pencil
def exibir_imgopencv(tmpimg):

  #Importar a Imagem
  im = cv2.imread(tmpimg)
  im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

  imout = cv2.edgePreservingFilter(im, flags=cv2.RECURS_FILTER);
  cv2.imwrite("edge-preserving-recursive-filter.jpg", imout);

  imout = cv2.edgePreservingFilter(im, flags=cv2.NORMCONV_FILTER);
  cv2.imwrite("edge-preserving-normalized-convolution-filter.jpg", imout);

  # Filtro de aprimoramento de detalhes
  imout = cv2.detailEnhance(im);
  cv2.imwrite("detail-enhance.jpg", imout);

  # Filtro de desenho a lápis
  imout_gray, imout = cv2.pencilSketch(im, sigma_s=60, sigma_r=0.07, shade_factor=0.05);
  cv2.imwrite("pencil-sketch.jpg", imout_gray);
  cv2.imwrite("pencil-sketch-color.jpg", imout);

  # Filtro de estilização
  cv2.stylization(im,imout);
  cv2.imwrite("stylization.jpg", imout);

  #Mostra a imagem com matplotlib
  plt.imshow(imout)
  plt.show()  

  return

#Aplica efeito de desenho a lápis a uma imagem RGB
def exibir_imgrender(img_rgb):
  #Converta a imagem colorida em tons de cinza.
  #Inverta a imagem em tons de cinza para obter um negativo.
  #Aplique um desfoque gaussiano ao negativo da etapa 2.
  img_rgb = cv2.imread(img_rgb)
  img_rgb = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)
  img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
  img_blur = cv2.GaussianBlur(img_gray, (21, 21), 0, 0)
  img_blend = cv2.divide(img_gray, img_blur, scale=256)

  #Imagem que servirá de fundo para dar o efeito de pincel a mão.
  canvas = cv2.imread('/content/planodefundopincel.jpg', cv2.CV_8UC1)
  
  #Redimensionando para o mesmo tamanho da imagem de entrada
  canvas = cv2.resize(canvas, (img_blend.shape[1], img_blend.shape[0]))

  ##Combine a imagem em tons de cinza da etapa 1 com o negativo desfocado da etapa 3 usando um desvio de cor.
  img_blend = cv2.multiply(img_blend, canvas, scale=1. / 256)
  imout = cv2.cvtColor(img_blend, cv2.COLOR_GRAY2RGB)

  #Mostra a imagem com matplotlib
  plt.imshow(imout)
  plt.show()  

  return 
```
